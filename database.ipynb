```python
import sqlite3
```

```python
connection = sqlite3.connect('database.db',timeout=10)
cursor = connection.cursor()
```

```python
cursor.execute("DROP TABLE IF EXISTS history")
cursor.execute("""
CREATE TABLE IF NOT EXISTS history(
    id INTEGER PRIMARY KEY,
    word TEXT,
    meaning TEXT
);
""")
```

```python
connection.commit()
connection.close()
```

```python
print("DataBase is ready")
```
